# Get the Data

```python
# download_dataset.py
import os
import zipfile
import shutil
from pathlib import Path

def setup_lgg_dataset():
    """Download and setup LGG MRI segmentation dataset"""
    
    # Create directories
    os.makedirs('data', exist_ok=True)
    os.makedirs('data/lgg-mri-segmentation', exist_ok=True)
    
    print("ðŸ“¥ Downloading LGG MRI segmentation dataset...")
    
    # Download using kaggle CLI
    os.system('kaggle datasets download -d mateuszbuda/lgg-mri-segmentation')
    
    # Extract the dataset
    print("ï¿½ï¿½ Extracting dataset...")
    with zipfile.ZipFile('lgg-mri-segmentation.zip', 'r') as zip_ref:
        zip_ref.extractall('data/lgg-mri-segmentation')
    
    # Clean up zip file
    os.remove('lgg-mri-segmentation.zip')
    
    print("âœ… Dataset downloaded and extracted successfully!")
    
    # Show dataset structure
    print("\nðŸ“ Dataset structure:")
    for root, dirs, files in os.walk('data/lgg-mri-segmentation'):
        level = root.replace('data/lgg-mri-segmentation', '').count(os.sep)
        indent = ' ' * 2 * level
        print(f"{indent}{os.path.basename(root)}/")
        subindent = ' ' * 2 * (level + 1)
        for file in files[:5]:  # Show first 5 files
            print(f"{subindent}{file}")
        if len(files) > 5:
            print(f"{subindent}... and {len(files) - 5} more files")
setup_lgg_dataset()
```
